```python
mean = np.mean(df['km'])
std_dev = np.std(df['km'])

z_score = (df['km'] - mean)/std_dev
outlier = df['km'][np.abs(z_score) >2]
outlier
```

```python
### Interquartile Range (IQR) method
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filter_df = df[(df['value'] >= lower_bound) & (df['value']<=upper_bound)]
```

```python
test_df = pd.DataFrame({'cost':[10,12,13,14,13,100,12,15,9]})
plt.boxplot(test_df)
```

```python
Q1 = df['km'].quantile(0.25)
Q3 = df['km'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filter_df = df[(df['km'] >= lower_bound) & (df['km']<=upper_bound)]
filter_df
```

```python
## Zscore 
mean = df['km'].mean()
std = df['km'].std()

threshold = 3 # user defined threshold value

filtered_df = df[(np.abs((df['km'] - mean) / std) < threshold)]
```
