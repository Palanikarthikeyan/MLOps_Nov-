```python
# Tasks
# ----------
# 1. load emp.csv file content to pandas  -> df
# 2. get info about loaded DataFrame
# 3. List out row index , column index
# 4. display df shape
# 5. Test - any duplicated values are exists -> drop duplicated updated to input dataframe
# 6. Test - any NaN value is exists => drop NaN index based and Column based
# 7. Test - any NaN ->replace row/column 
# 8. Rename some selective columns - modified input data frame(ex: edept =>DEPT 'ecost' ->Amount)
# 9. Drop some selective columns
# 10. Write DataFrame to External File
----------------------------------------------------------------
```

```python
# 1. load emp.csv file content to pandas  -> df

df_csv = pd.read_csv('emp.csv')


# 2. get info about loaded DataFrame
df_csv.info()

# 3. List out row index , column index
print('Row index:',df_csv.index)
print('List of columns:',df_csv.columns)

# 4. Display shape of the data - Row X Col
print(df_csv.shape)
```

```python
# # 5. Test - any duplicated values are exists -> drop duplicated updated to input dataframe
df_csv.duplicated()
df_csv.duplicated().sum()
df_csv.drop_duplicates(inplace=True)
```

```python
# 6. Test - any NaN value is exists => drop NaN index based and Column based
df_csv.isnull().sum()

# 7. Test - any NaN ->replace row/column 
df_csv.dropna() # default rowbased 
df_csv.dropna(axis=1) # columnbased 
df_csv.fillna('sampledata') # replace NaN ->sampledata
df_csv['edept'].fillna('commonDept')
```

```python
# 8. Rename some selective columns - modified input data frame(ex: edept =>DEPT 'ecost' ->Amount)
print(df_csv.columns)
df_csv.rename(columns={'edept':'DEPT','ecost':'Amount'},inplace=True)
print(df_csv.columns)
```

```python
# 9. Drop some selective columns
df_csv.drop(columns=['eid','ecity'])
```

```python
df_csv.drop(columns=['eid','ecity'],inplace=True)
```

```python
# Transform(inside the db) - inside the DB Via SQL not using python programming 
conn.execute("""create table product_final as 
                SELECT pid,
                      pname,
                      pcost, 
                      CASE 
                        when pname == 'pC' THEN pcost * 18
                      ELSE pcost 
                   END AS pcost
                  FROM product_raw;
                """)
conn.commit()
```

```python
# Testing
for var in conn.execute("select *from product_final"):
    print(var)
```
